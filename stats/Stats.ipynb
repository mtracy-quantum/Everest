```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('member_export.csv', encoding="ISO-8859-1", low_memory=False)
```

```python
df.info()
```

```python
# The dataset has Mallory and Irvine, so remove them
# this command removes "hired" as Sherpa deaths happen for different reasons
# Also remove misc items that are not relevant
north_climbs = df[(df['hired']==False) & (df['route1'].str[0]=='N') & (df['mperhighpt']>=8000) &
                 (df['fname']!='Andrew Comyn (Sandy)') & (df['fname']!='George Herbert Leigh') & 
                 (df['mski'] ==False) & (df['route1']!='N Col (training, recon)')]
```

```python
# a dataset for the Sherpa

north_climbs_h = df[ (df['route1'].str[0]=='N') & (df['mperhighpt']>=8000) &
                 (df['fname']!='Andrew Comyn (Sandy)') & (df['fname']!='George Herbert Leigh') & 
                 (df['mski'] ==False) & (df['route1']!='N Col (training, recon)')]
```

```python
# just use the columns we are interested in.  "msuccess" is "member success" and means the climber reached the summit
north_climbs = north_climbs[['expid', 'fname', 'lname', 'sex', 'yob', 'calcage', 'hired', 'sherpa', 'msuccess', 'mo2used', 'mperhighpt', 'death', 'deathhgtm',  'route1']]
# just use the columns we are interested in.  "msuccess" is "member success" and means the climber reached the summit
north_climbs_h = north_climbs_h[['expid', 'fname', 'lname', 'sex', 'yob', 'calcage', 'hired', 'sherpa', 'msuccess', 'mo2used', 'mperhighpt', 'death', 'deathhgtm',  'route1']]
```

```python
# Look at what happens when you turn around between First and Second Step

# First Step
north_climbs[(north_climbs['mperhighpt'] >=8500) & (north_climbs['mperhighpt'] <=8600)].groupby(by=['death']).count()
```

 Assume nothing other than that they reached the "death zone" and died
 $$P(Summit|Reached\, 8000m)$$

```python
death_zone = north_climbs[ (north_climbs['death']==True)]
death_zone.groupby(by=['msuccess']).count()
```

```python
(33)/(33+24)
```

$P(Summit|Reached\, 8000m) = \frac{33}{33+24} = 58\%$

Using only that we know they reached the location of the ice axe (8840m) -- it does not assume it was either the ascent or descent



```python
north_climbs[(north_climbs['death'] == True) & (north_climbs['mperhighpt'] >=8440)].groupby(by=['msuccess']).count()
```

```python
(33/(16+33))
```

$P(Summit|Reached\, 8440m) = \frac{33}{33+16} = 67\%$

And the theory that they died at the ice axe site on the ascent which I will calculate at the percentage of people who die on the ascent between 8000 and 8500 meters

```python
ascent_below_first_step = north_climbs[(north_climbs['death'] == True) & (north_climbs['mperhighpt'] <= 8500 ) & (north_climbs['mperhighpt'] == north_climbs['deathhgtm'])]
ascent_below_first_step.groupby(by=['mperhighpt']).count()  # we really just need the total count, but this will break out where they died
```

A total of 9 people out of 2835 climbs died on the ascent below the First Step

Next we look at death on descent after having reached at least the ice axe location

```python
death_on_descent = north_climbs[(north_climbs['death'] == True)  & (north_climbs['mperhighpt'] >=8440) & (north_climbs['mperhighpt'] > north_climbs['deathhgtm'])]
death_on_descent.groupby(by=['msuccess']).count()
```

```python
33/(33+11)
```

$P(Summit|Died\,on\,Descent) = \frac{35}{35+14} = 75\%$

```python
# Lets look at people who turn around between the Second and Third Steps
turn_around_between_second_third = north_climbs[(north_climbs['mperhighpt']>=8600) &(north_climbs['mperhighpt']<=8650)]
turn_around_between_second_third.groupby(by=['death']).count()
```

```python
# for the 149 climbers who turned around after reaching the Third Step, lets look at the results
# Note 1 person died above the third step on the ascent -- that is why it does not match the 150 above


third_step_turn_around = north_climbs[(north_climbs['mperhighpt']>=8650) & (north_climbs['msuccess']==False) & (north_climbs['mperhighpt'] > north_climbs['deathhgtm'])]
third_step_turn_around.groupby(by=['death']).count()
```

```python
# died on ascent above Third Step
north_climbs[(north_climbs['mperhighpt']>=8650) & (north_climbs['msuccess']==False) & (north_climbs['mperhighpt'] == north_climbs['deathhgtm'])]
```

Now compute
$$P(Summit|Reached\,Third\,Step\,and\,died\,on\,descent) $$

```python
at_least_third_step = north_climbs[(north_climbs['mperhighpt']>=8650) & (north_climbs['death']==True) & (north_climbs['mperhighpt'] > north_climbs['deathhgtm'])]
at_least_third_step.groupby(by=['msuccess']).count()
```

```python
33/(33+5)
```


$P(Summit|Reached\,Third\,Step\,and\,died\,on\,descent)  = \frac{33}{33+5} = 87\%$  

```python
# verify that including 'hired' does not meaningful chance the results.
# I leave out the hired because their deaths are largely due to the clients mistakes, 
# they carry much larger loads, and use much less oxygen.
# All of their names are listed at the end of this analysis
at_least_third_step_h = north_climbs_h[(north_climbs_h['mperhighpt']>=8650) & (north_climbs_h['death']==True) & (north_climbs_h['mperhighpt'] > north_climbs_h['deathhgtm'])]
at_least_third_step_h.groupby(by=['msuccess']).count()
```
